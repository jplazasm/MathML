```python
from datetime import datetime, timedelta
import pandas as pd
```

```python
#Crear un dataframe con todas las fechas del año 2021

inicio = datetime(2021,1,1);
fin   = datetime(2021,12,31);

lista_fechas = [(inicio + timedelta(days=d)).strftime("%Y-%m-%d")
                    for d in range((fin - inicio).days + 1)] 
Date = pd.DataFrame(lista_fechas, columns = ["Date"])
Date["Date"] = Date["Date"].astype("datetime64[ns]")
```

```python
bit2021 = pd.read_csv(r"C:\Users\felip\OneDrive\Documentos\MathML\Taller 1\Datos\bit2021.csv")
gold2021 = pd.read_csv(r"C:\Users\felip\OneDrive\Documentos\MathML\Taller 1\Datos\gold2021.csv")
oil2021 = pd.read_csv(r"C:\Users\felip\OneDrive\Documentos\MathML\Taller 1\Datos\oil2021.csv")
silver2021 = pd.read_csv(r"C:\Users\felip\OneDrive\Documentos\MathML\Taller 1\Datos\silver2021.csv")
```

```python
def clean_dfs(df):
    df = df.drop("Unnamed: 0",1)
    df.columns = ["Date", "Price"]

    df["Date"] = df["Date"].astype("datetime64[ns]")
    df = df.sort_values(by=['Date'], ascending = True)

    df["Price"] = df["Price"].astype("string")
    df["Price"] = df["Price"].str.replace(',', '.') 
    df["Price"] = df["Price"].astype("float64")
    return df

def merge_interpolate_dfs():
    dfs_prices = [bit2021, gold2021, oil2021, silver2021]
    clean_dfs_prices = list(map(clean_dfs, dfs_prices))

    prices = Date
    for df in clean_dfs_prices:
        prices = prices.merge(df, on="Date", how="left")
        
    columns_prices = ["Date","PriceBitcoin", "PriceGold", "PriceOil", "PriceSilver"]
    prices.columns = columns_prices
    prices.iloc[:,1:5] = prices.iloc[:,1:5].interpolate(method='linear', axis=0) 
    prices = prices.fillna(prices.mean())
    prices.iloc[:,1:5] = prices.iloc[:,1:5].round(2)
    return prices
```

```python
prices = merge_interpolate_dfs()
prices.to_csv(r"C:\Users\felip\OneDrive\Documentos\MathML\Taller 1\prices2021.csv")
prices
```
